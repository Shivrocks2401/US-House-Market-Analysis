```python
import pandas as pd
import numpy as np
import datetime
```

# Reading all datasets from different sources

```python

housePriceIndex= pd.read_csv("datasets for task/USA HOUSE PRICE INDEX.csv")
#housePriceIndex=
FinalDataset=housePriceIndex
FinalDataset["housePriceIndex"]=housePriceIndex["CSUSHPISA"]
del FinalDataset["CSUSHPISA"]
```

```python
FinalDataset["DATE"]=pd.to_datetime(FinalDataset["DATE"])
FinalDataset=FinalDataset.resample("A", on="DATE").mean()
FinalDataset
```

```python
############RESIDENTIAL DEVELOPMENT INVESTMENT################################################
columnsForRDI=["Date","TotalConstruction",'TotalPrivateConstruction','TotalPublicConstruction']
rdi = pd.read_excel("datasets for task/residential development investment.xls",header=0)
rdi=rdi[columnsForRDI]
rdi[["mm", "DATE"]] = rdi["Date"].str.split("-", expand=True)
rdi[["DATE"]]=rdi["DATE"].str[0:2]
rdi=rdi.astype({"DATE":"int32"})
actualYear= [2000+x if x<=20 else 1900+x for x in rdi["DATE"]]
rdi[["DATE"]]=actualYear
del rdi["Date"]
del rdi["mm"]
rdi
```

```python
rdi=rdi[::-1]
rdi["DATE"]=rdi["DATE"].astype("str")
strings=[]
for i in range(len(rdi["DATE"])):
    strings.append("12-31")
    
rdi["DATE"]=rdi["DATE"].str.cat(strings,sep="-")

rdi["DATE"]=pd.to_datetime(rdi["DATE"])
rdi
```

```python
rdi=rdi.resample("A", on="DATE").mean()
rdi
```

```python
#############Average Income########################################
avInc=pd.read_csv("datasets for task/average(median household income).csv")
avInc[["DATE"]]=avInc["DATE"].str[0:4]
avInc[["averageIncome"]]=avInc["MEHOINUSA672N"]
del avInc["MEHOINUSA672N"]
avInc["DATE"]=avInc["DATE"].astype("str")
#avInc=avInc.set_index("DATE")
strings=[]
for i in range(len(avInc["DATE"])):
    strings.append("12-31")
    
avInc["DATE"]=avInc["DATE"].str.cat(strings,sep="-")

avInc["DATE"]=pd.to_datetime(avInc["DATE"])
avInc=avInc.set_index("DATE")
avInc
```

```python
################Consumer Price Index############################
cpi= pd.read_csv("datasets for task/consumer price index second by world bank.csv")
cpi=cpi.set_index("Country Name").T[1:]
cpi=cpi["United States"]
cpi=pd.DataFrame(cpi)
indexes=[]
for i in cpi.index:
    indexes.append(i)

for i in range(len(indexes)):
    indexes[i]=str(indexes[i])+"-12-31"
indexes
cpi["DATE"]=indexes
cpi["DATE"]=pd.to_datetime(cpi["DATE"])
cpi=cpi.set_index("DATE")
cpi["consumerPriceIndex"]=cpi["United States"]
del cpi["United States"]
cpi
```

```python
###################Gross Value Added#########################
gva= pd.read_csv("datasets for task/gross value added US.csv")
gva=gva.iloc[7::8, :]
gva=gva.loc[:,('Year', 'Value')]
gva["DATE"]=gva["Year"].astype("str")
strings=[]
for i in range(len(gva["Year"])):
    strings.append("12-31")
    
gva["DATE"]=gva["DATE"].str.cat(strings,sep="-")

gva["DATE"]=pd.to_datetime(gva["DATE"])
gva=gva.set_index("DATE")
gva=gva[::-1]
gva["grossValueAdded"]=gva["Value"]
del gva["Value"]
del gva["Year"]
gva
```

```python
#############housing construction starts####################
dwellingStart=pd.read_excel("datasets for task/housing unit starts_cust.xls")
dwellingStart["DATE"]=dwellingStart["Month"]

dwellingStart["DATE"]=pd.to_datetime(dwellingStart["DATE"])

dwellingStart=dwellingStart.resample("A", on="DATE").sum()
dwellingStart["dwellingsStarted"]=dwellingStart["Total"]
del dwellingStart["Total"]
#del dwellingStart["Month"]
dwellingStart.head()
```

```python
#############housing construction authorized in permit area####################

dwellingauthorizedPermit=pd.read_excel("datasets for task/housing units authorized in permits_cust.xls")
dwellingauthorizedPermit["DATE"]=dwellingauthorizedPermit["Month"]

dwellingauthorizedPermit["DATE"]=pd.to_datetime(dwellingauthorizedPermit["DATE"])

dwellingauthorizedPermit=dwellingauthorizedPermit.resample("A", on="DATE").sum()
dwellingauthorizedPermit["dwellingsAuthorizedPermit"]=dwellingauthorizedPermit["Total"]
del dwellingauthorizedPermit["Total"]
dwellingauthorizedPermit.head()
```

```python
#############housing construction completed in permit area####################

dwellingCompletedPermit=pd.read_excel("datasets for task/housing units completed in permit areas.xls")
dwellingCompletedPermit["DATE"]=dwellingCompletedPermit["Month"]

dwellingCompletedPermit["DATE"]=pd.to_datetime(dwellingCompletedPermit["DATE"])

dwellingCompletedPermit=dwellingCompletedPermit.resample("A", on="DATE").sum()
dwellingCompletedPermit["dwellingsCompletedPermit"]=dwellingCompletedPermit["Total"]
del dwellingCompletedPermit["Total"]
dwellingCompletedPermit.head()
```

```python
#############housing construction starts in permit area####################
dwellingStartPermit=pd.read_excel("datasets for task/housing units starts in permit areas_cust.xls")
dwellingStartPermit["DATE"]=dwellingStartPermit["Month"]

dwellingStartPermit["DATE"]=pd.to_datetime(dwellingStartPermit["DATE"])


dwellingStartPermit=dwellingStartPermit.resample("A", on="DATE").sum()
dwellingStartPermit["dwellingsStartedPermit"]=dwellingStartPermit["Total"]
del dwellingStartPermit["Total"]
dwellingStartPermit.head()
```

```python
#############housing construction in process in permit area####################

dwellingprogressPermit=pd.read_excel("datasets for task/housing units under construction in permit areas.xls")
dwellingprogressPermit["DATE"]=dwellingprogressPermit["Month"]
dwellingprogressPermit["DATE"]=pd.to_datetime(dwellingprogressPermit["DATE"])

dwellingprogressPermit=dwellingprogressPermit.resample("A", on="DATE").sum()
dwellingprogressPermit["dwellingsInProgressPermit"]=dwellingprogressPermit["Total"]
del dwellingprogressPermit["Total"]
dwellingprogressPermit.head()
```

```python
#############Population density####################
populationDensity=pd.read_csv("datasets for task/united-states-population-2021-01-30.csv")
populationDensity[["DATE"]]=pd.to_datetime(populationDensity["date"])
populationDensity["DATE"]=populationDensity['DATE'].dt.strftime('%Y-%m-%d')
populationDensity["DATE"]=pd.to_datetime(populationDensity["DATE"])


populationDensity=populationDensity.set_index("DATE")
del populationDensity["date"]
populationDensity.head()
```

```python
################Consumer Confidence Index################################
cci=pd.read_csv("datasets for task/USA Consumer Confidence Index.csv")
cci[["DATE"]]=pd.to_datetime(cci["TIME"])
cci["consumerConfidenceIndex"]=cci["Value"]
del cci["Value"]
cci=cci.resample("A", on="DATE").mean()
cci.head()
```

```python
################Gross disposable household income################################
gdhi=pd.read_csv("datasets for task/USA gross disposable household income.csv")
gdhi["TIME"]= gdhi["TIME"].astype("str")
gdhi["TIME"]= gdhi["TIME"]
gdhi["grossDisposableHouseholdIncome"]=gdhi["Value"]
del gdhi["Value"]

gdhi["DATE"]=pd.to_datetime(gdhi["TIME"])
gdhi=gdhi.resample("A", on="DATE").mean()
#del gdhi["TIME"]
gdhi.head()
```

```python
################Unemployment rate################################
ur=pd.read_csv("datasets for task/USA Unemployment rate.csv")
ur["DATE"]=pd.to_datetime(ur["DATE"])
ur["unemploymentRate"]=ur["UNRATE"]
del ur["UNRATE"]
ur=ur.resample("A", on="DATE").mean()
ur.head()
```

```python
###############HomeOwnership Rate#################################
hor=pd.read_csv("datasets for task/homeownership rate.csv")
hor["Date"]=pd.to_datetime(hor["Date"])
hor["Homeownership(in %)"] =hor["Homeownership(in %)"].str[0:5]
hor["Homeownership(in %)"]=hor["Homeownership(in %)"].astype("float64")
hor["DATE"]=hor["Date"]
hor["DATE"]=pd.to_datetime(hor["DATE"])
hor=hor.resample("A", on="DATE").mean()
hor
```

```python
###########Interest Rate##############################
ir=pd.read_csv("datasets for task/INTDSRUSM193N.csv")
ir["DATE"]=pd.to_datetime(ir["DATE"])
ir=ir.resample("A", on="DATE").mean()
ir
```

```python
###########Mortgage Rate##############################
mr=pd.read_csv("datasets for task/MORTGAGE30US.csv")
mr["DATE"]=pd.to_datetime(mr["DATE"])
mr=mr.resample("A", on="DATE").mean()
mr
```

## Merging All Dataset

```python
df=[FinalDataset,
    rdi,
    avInc,
    cpi,
    gva,
    dwellingauthorizedPermit,
    dwellingCompletedPermit,
    dwellingprogressPermit,
    dwellingStart,
    dwellingStartPermit,
   populationDensity,
    cci,
    gdhi,
   ur,hor,ir,mr]
```

```python
import functools 
df = functools.reduce(lambda  left,right: pd.merge(left,right,on=["DATE"],
                                            how='outer'), df)

```
